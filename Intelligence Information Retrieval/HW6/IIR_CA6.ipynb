```python

# ==================================================================
#@title STEP 1: Document Parsing (5 points)
# ==================================================================
print("\n" + "=" * 60)
print("ğŸ“„ STEP 1: Document Parsing")
print("=" * 60)

def parse_pdfs(folder_path: str) -> pd.DataFrame:
    """Parse PDFs and extract first page text with metadata"""
    pdf_files = list(Path(folder_path).glob("*.pdf"))
    print(f"Found {len(pdf_files)} PDF files")

    documents = []
    for pdf_path in pdf_files:
        try:
            doc_id = pdf_path.stem.replace("Copy of ", "").strip()
            doc = pymupdf.open(pdf_path)
            text = doc[0].get_text()
            lines = text.strip().split('\n')

            documents.append({
                "doc_id": doc_id,
                "title": lines[0][:200] if lines else "Unknown",
                "first_page_text": text,
                "text_length": len(text)
            })
            doc.close()
        except Exception as e:
            print(f"Error: {pdf_path.name} - {e}")

    return pd.DataFrame(documents)

# Parse documents
docs_df = parse_pdfs(PAPERS_FOLDER)

print(f"\nâœ… Parsed {len(docs_df)} documents")
print("\nğŸ“‹ DataFrame Preview:")
print(docs_df[['doc_id', 'title', 'text_length']].head(10))

print(f"\nğŸ“Š Statistics:")
print(f"   Total documents: {len(docs_df)}")
print(f"   Avg text length: {docs_df['text_length'].mean():.0f} chars")
```
