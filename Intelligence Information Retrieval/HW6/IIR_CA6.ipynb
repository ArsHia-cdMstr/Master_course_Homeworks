```python
"""
==================================================================
#@title Part 1: Document Parsing + Vector Database + Semantic Search
==================================================================
Uses the Avalai client from previous cell
"""

import pymupdf
import pandas as pd
import lancedb
from pathlib import Path
from sentence_transformers import SentenceTransformer
from google.colab import drive

# ------------------------------------------------------------------
# Config
# ------------------------------------------------------------------
PAPERS_FOLDER = "/content/drive/My Drive/Colab Notebooks/Master_course_Homeworks/Intelligence Information Retrieval/HW6/Data/"
DB_PATH = "/content/papers_db"
EMBEDDING_MODEL = "BAAI/bge-small-en-v1.5"

# ------------------------------------------------------------------
# Mount Drive (if not already mounted)
# ------------------------------------------------------------------
try:
    drive.mount('/content/drive')
except:
    print("Drive already mounted")

# ------------------------------------------------------------------
# Document Parser
# ------------------------------------------------------------------
def parse_pdfs(folder_path: str) -> pd.DataFrame:
    """
    Parse all PDFs in folder and extract first page text.

    Parameters
    ----------
    folder_path : str
        Path to folder containing PDFs

    Returns
    -------
    pd.DataFrame
        DataFrame with doc_id, title, first_page_text, text_length
    """
    pdf_files = list(Path(folder_path).glob("*.pdf"))
    print(f"Found {len(pdf_files)} PDF files")

    documents = []
    for pdf_path in pdf_files:
        try:
            doc = pymupdf.open(pdf_path)
            text = doc[0].get_text()
            lines = text.strip().split('\n')

            documents.append({
                "doc_id": pdf_path.stem,
                "title": lines[0][:200] if lines else "Unknown",
                "first_page_text": text,
                "text_length": len(text)
            })
            doc.close()

        except Exception as e:
            print(f"Error parsing {pdf_path.name}: {e}")

    return pd.DataFrame(documents)

# ------------------------------------------------------------------
# Vector Database Manager
# ------------------------------------------------------------------
class VectorDB:
    """Simple wrapper for LanceDB operations."""

    def __init__(self, db_path: str, embedding_model_name: str):
        """
        Initialize vector database.

        Parameters
        ----------
        db_path : str
            Path to store database
        embedding_model_name : str
            Name of sentence-transformers model
        """
        print(f"Loading embedding model: {embedding_model_name}")
        self.model = SentenceTransformer(embedding_model_name)
        self.db = lancedb.connect(db_path)
        self.table = None

    def create_index(self, docs_df: pd.DataFrame, text_column: str = "first_page_text"):
        """
        Create vector index from documents.

        Parameters
        ----------
        docs_df : pd.DataFrame
            DataFrame with documents
        text_column : str
            Column name containing text to embed
        """
        print("Generating embeddings...")
        texts = docs_df[text_column].tolist()
        embeddings = self.model.encode(texts, show_progress_bar=True)

        docs_df = docs_df.copy()
        docs_df['vector'] = embeddings.tolist()

        print("Creating database table...")
        self.table = self.db.create_table("papers", docs_df, mode="overwrite")
        print(f"Index created with {self.table.count_rows()} documents")

    def search(self, query: str, top_k: int = 5) -> pd.DataFrame:
        """
        Search for similar documents.

        Parameters
        ----------
        query : str
            Search query
        top_k : int
            Number of results

        Returns
        -------
        pd.DataFrame
            Search results with doc_id, title, distance
        """
        query_vector = self.model.encode(query).tolist()
        results = self.table.search(query_vector).limit(top_k).to_pandas()
        return results[['doc_id', 'title', 'text_length', '_distance']]

# ------------------------------------------------------------------
# Evaluation Functions
# ------------------------------------------------------------------
def calculate_p_at_k(vector_db: VectorDB, users_data: list, relevant_docs: dict, k: int = 1) -> float:
    """
    Calculate Precision at K.

    Parameters
    ----------
    vector_db : VectorDB
        Vector database instance
    users_data : list
        List of user dictionaries with information_need
    relevant_docs : dict
        Mapping of user_id to list of relevant doc_ids
    k : int
        K value for P@K

    Returns
    -------
    float
        P@K score
    """
    correct = 0
    total = 0

    for user in users_data:
        user_id = user.get('user_id', user.get('id'))
        query = user.get('information_need', user.get('query', ''))

        if not query or user_id not in relevant_docs:
            continue

        results = vector_db.search(query, top_k=k)
        retrieved_docs = results['doc_id'].tolist()

        # Check if any retrieved doc is relevant
        for doc in retrieved_docs:
            if doc in relevant_docs[user_id]:
                correct += 1
                break

        total += 1

    return correct / total if total > 0 else 0

# ------------------------------------------------------------------
# Main Execution
# ------------------------------------------------------------------

# 1. Parse documents
print("=" * 60)
print("Step 1: Parsing PDFs")
print("=" * 60)
docs_df = parse_pdfs(PAPERS_FOLDER)
print(f"\nParsed {len(docs_df)} documents")
print(docs_df[['doc_id', 'title', 'text_length']].head())

# 2. Create vector database
print("\n" + "=" * 60)
print("Step 2: Creating Vector Database")
print("=" * 60)
vector_db = VectorDB(DB_PATH, EMBEDDING_MODEL)
vector_db.create_index(docs_df)

# 3. Test semantic search
print("\n" + "=" * 60)
print("Step 3: Testing Semantic Search")
print("=" * 60)
test_query = "machine learning for text classification"
print(f"Query: {test_query}\n")
results = vector_db.search(test_query, top_k=5)
print("Top 5 Results:")
print(results)

# 4. Test with user data (if loaded from previous cell)
# 4. Test with user data (if loaded from previous cell)
print("\n" + "=" * 60)
print("Step 4: Testing with User Queries")
print("=" * 60)

try:
    for i, user in enumerate(users_data[:3]):
        # Ø¯Ø±ÛŒØ§ÙØª query Ø¨Ø§ Ø¨Ø±Ø±Ø³ÛŒ Ù†ÙˆØ¹ Ø¯Ø§Ø¯Ù‡
        query = user.get('information_need', user.get('query', ''))

        # Ø§Ú¯Ø± query Ø¯ÛŒÚ©Ø´Ù†Ø±ÛŒ ÛŒØ§ Ù„ÛŒØ³Øª Ø¨ÙˆØ¯ØŒ Ø¢Ù† Ø±Ø§ Ø¨Ù‡ Ø±Ø´ØªÙ‡ ØªØ¨Ø¯ÛŒÙ„ Ú©Ù†ÛŒØ¯
        if isinstance(query, dict):
            query = str(query)  # ÛŒØ§ query.get('text', '') Ø§Ú¯Ø± Ú©Ù„ÛŒØ¯ Ø®Ø§ØµÛŒ Ø¯Ø§Ø±Ø¯
        elif isinstance(query, list):
            query = ' '.join(map(str, query))

        # Ù†Ù…Ø§ÛŒØ´ query (Ø¨Ø§ Ù…Ø­Ø¯ÙˆØ¯ÛŒØª 60 Ú©Ø§Ø±Ø§Ú©ØªØ±)
        display_query = query[:60] if len(query) > 60 else query
        print(f"\nUser {i+1}: {display_query}...")

        # Ø¬Ø³ØªØ¬Ùˆ
        results = vector_db.search(query, top_k=3)
        for _, row in results.iterrows():
            print(f"  - {row['doc_id']} (dist: {row['_distance']:.4f})")

except NameError:
    print("users_data not found. Load it first from the JSON file.")
except Exception as e:
    print(f"Error: {e}")
    print(f"Sample user data structure: {users_data[0] if users_data else 'Empty'}")

# 5. Statistics
print("\n" + "=" * 60)
print("Database Statistics")
print("=" * 60)
print(f"Total documents: {len(docs_df)}")
print(f"Avg text length: {docs_df['text_length'].mean():.0f} chars")
print(f"Embedding dimension: {len(vector_db.model.encode('test'))}")
```
