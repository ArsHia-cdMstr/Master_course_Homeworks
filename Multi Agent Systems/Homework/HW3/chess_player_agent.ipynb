```python
#@title 0.3 - fen_explainer.py
# ============================================
# Demonstrates FEN parsing and explanation
# using python-chess library
# ============================================

import chess


def explain_fen(fen_string: str) -> dict:
    """
    Parse a FEN string and return a human-readable
    explanation of all 6 components.

    Args:
        fen_string: A valid FEN string

    Returns:
        Dictionary with explanation of each FEN component
    """
    board = chess.Board(fen_string)
    parts = fen_string.split(" ")

    explanation = {}

    # --- Part 1: Piece Placement ---
    explanation["piece_placement"] = parts[0]
    explanation["piece_placement_note"] = (
        "Rows separated by '/'. "
        "Uppercase = White, lowercase = Black. "
        "Numbers = empty squares."
    )

    # --- Part 2: Active Color ---
    explanation["active_color"] = "White" if parts[1] == "w" else "Black"

    # --- Part 3: Castling Rights ---
    castling = parts[2]
    rights = []
    if "K" in castling:
        rights.append("White Kingside")
    if "Q" in castling:
        rights.append("White Queenside")
    if "k" in castling:
        rights.append("Black Kingside")
    if "q" in castling:
        rights.append("Black Queenside")
    if castling == "-":
        rights.append("None")
    explanation["castling_rights"] = rights

    # --- Part 4: En Passant Target ---
    explanation["en_passant"] = parts[3] if parts[3] != "-" else "None"

    # --- Part 5: Halfmove Clock ---
    explanation["halfmove_clock"] = int(parts[4])
    explanation["halfmove_note"] = (
        "Moves since last pawn advance or capture. "
        "Used for 50-move draw rule."
    )

    # --- Part 6: Fullmove Number ---
    explanation["fullmove_number"] = int(parts[5])

    # --- Extra: Board visualization ---
    explanation["board_visual"] = str(board)

    # --- Extra: Legal moves count ---
    explanation["legal_moves_count"] = board.legal_moves.count()

    # --- Extra: Game status ---
    explanation["is_check"] = board.is_check()
    explanation["is_checkmate"] = board.is_checkmate()
    explanation["is_stalemate"] = board.is_stalemate()

    return explanation


# ============================================
# Example 1: Starting position after 1. e4
# ============================================
fen_1 = "rnbqkbnr/pppppppp/8/8/4P3/8/PPPP1PPP/RNBQKBNR b KQkq e3 0 1"

print("=" * 50)
print("EXAMPLE 1: After 1. e4")
print("=" * 50)
result = explain_fen(fen_1)

for key, value in result.items():
    print(f"  {key}: {value}")

# ============================================
# Example 2: A mid-game position
# ============================================
fen_2 = "r1bqk2r/pppp1ppp/2n2n2/2b1p3/2B1P3/5N2/PPPP1PPP/RNBQK2R w KQkq - 4 4"

print("\n" + "=" * 50)
print("EXAMPLE 2: Italian Game (4 moves in)")
print("=" * 50)
result = explain_fen(fen_2)

for key, value in result.items():
    print(f"  {key}: {value}")

# ============================================
# Example 3: Checkmate position
# ============================================
fen_3 = "rnb1kbnr/pppp1ppp/8/4p3/6Pq/5P2/PPPPP2P/RNBQKBNR w KQkq - 1 3"

print("\n" + "=" * 50)
print("EXAMPLE 3: Fool's Mate (checkmate!)")
print("=" * 50)
result = explain_fen(fen_3)

for key, value in result.items():
    print(f"  {key}: {value}")
```

```python
#@title  FEN Explainer with Graphical Board
# ============================================
# Visual FEN explanation with chess board
# ============================================

import chess
import chess.svg
from IPython.display import display, SVG, HTML  # â† Ø§ÛŒÙ† Ø®Ø· Ù…Ù‡Ù…Ù‡!


def explain_fen_graphical(fen_string: str, last_move: str = None):
    """
    Parse and display FEN with graphical board.

    Args:
        fen_string: Valid FEN string
        last_move: Optional last move in UCI format to highlight
    """
    board = chess.Board(fen_string)
    parts = fen_string.split(" ")

    # --- Text explanation ---
    print("ğŸ“Œ FEN EXPLANATION")
    print("=" * 40)

    print(f"1ï¸âƒ£ Piece placement:")
    print(f"    {parts[0]}")
    print(f"   Rows separated by '/', uppercase = White, lowercase = Black\n")

    print(f"2ï¸âƒ£ Side to move:")
    print(f"    {'White âšª' if parts[1] == 'w' else 'Black âš«'}\n")

    print(f"3ï¸âƒ£ Castling rights:")
    castling = parts[2]
    if castling == "-":
        print(f"    None")
    else:
        rights = []
        if "K" in castling: rights.append("White O-O")
        if "Q" in castling: rights.append("White O-O-O")
        if "k" in castling: rights.append("Black O-O")
        if "q" in castling: rights.append("Black O-O-O")
        print(f"    {', '.join(rights)}\n")

    print(f"4ï¸âƒ£ En passant target:")
    print(f"    {parts[3] if parts[3] != '-' else 'None'}\n")

    print(f"5ï¸âƒ£ Halfmove clock:")
    print(f"    {parts[4]} (50-move rule counter)\n")

    print(f"6ï¸âƒ£ Fullmove number:")
    print(f"    {parts[5]}\n")

    print(f"â™Ÿï¸ Game status:")
    print(f"   Check: {board.is_check()}")
    print(f"   Checkmate: {board.is_checkmate()}")
    print(f"   Stalemate: {board.is_stalemate()}")
    print(f"   Legal moves: {board.legal_moves.count()}")

    # --- Graphical board ---
    print(f"\nğŸ–¼ï¸ Board view:")

    if last_move:
        try:
            move = chess.Move.from_uci(last_move)
            display(SVG(chess.svg.board(board, lastmove=move, size=400)))
        except:
            display(SVG(chess.svg.board(board, size=400)))
    else:
        display(SVG(chess.svg.board(board, size=400)))

    return board


def show_move_transition(fen: str, move_uci: str):
    """
    Show before and after boards for a move.

    Args:
        fen: Starting position FEN
        move_uci: Move in UCI format (e.g., "e2e4")
    """
    board_before = chess.Board(fen)
    move = chess.Move.from_uci(move_uci)
    san_move = board_before.san(move)

    print("=" * 50)
    print(f"ğŸ¬ MOVE: {san_move} ({move_uci})")
    print("=" * 50)

    # Before
    print("\nğŸ“ BEFORE:")
    svg_before = chess.svg.board(
        board_before,
        size=350,
        arrows=[(move.from_square, move.to_square)],  # Arrow showing move
        fill={move.from_square: "#ffff0080"}  # Highlight source square
    )
    display(SVG(svg_before))

    # Apply move
    board_after = board_before.copy()
    board_after.push(move)

    # After
    print("\nğŸ“ AFTER:")
    svg_after = chess.svg.board(
        board_after,
        size=350,
        lastmove=move,  # Highlight the move
        fill={move.to_square: "#90EE9080"}  # Green highlight
    )
    display(SVG(svg_after))

    print(f"\nâœ… New FEN: {board_after.fen()}")
    print(f"   Turn: {'White âšª' if board_after.turn else 'Black âš«'}")
    print(f"   Check: {board_after.is_check()}")

    return board_after


# ============================================
# Test Examples
# ============================================

# Example 1: Explain FEN
print("ğŸ¯ EXAMPLE 1: Position after 1. e4")
print("=" * 50)
fen = "rnbqkbnr/pppppppp/8/8/4P3/8/PPPP1PPP/RNBQKBNR b KQkq e3 0 1"
explain_fen_graphical(fen)

# Example 2: Show a move
print("\n\nğŸ¯ EXAMPLE 2: Black plays c5 (Sicilian Defense)")
show_move_transition(fen, "c7c5")
```

```python
#@title 0.4 - move_validation.py
# ============================================
# Demonstrates how to validate and apply moves
# This is CRITICAL for the agent: LLMs often
# suggest illegal moves (hallucination)
# ============================================

import chess


def validate_and_apply_move(board: chess.Board, move_str: str) -> dict:
    """
    Validate a move string and apply it if legal.

    Supports multiple formats:
    - UCI format: "e2e4", "g1f3"
    - SAN format: "e4", "Nf3", "O-O"

    Args:
        board: Current chess board state
        move_str: The move string to validate

    Returns:
        Dictionary with validation result and new FEN
    """
    result = {
        "input_move": move_str,
        "is_valid": False,
        "applied": False,
        "new_fen": None,
        "error": None
    }

    # --- Try UCI format first (e.g., "e2e4") ---
    try:
        move = chess.Move.from_uci(move_str)
        if move in board.legal_moves:
            board.push(move)
            result["is_valid"] = True
            result["applied"] = True
            result["new_fen"] = board.fen()
            result["format_detected"] = "UCI"
            return result
        else:
            result["error"] = f"Move {move_str} is not legal in UCI format"
    except ValueError:
        pass

    # --- Try SAN format (e.g., "Nf3", "e4") ---
    try:
        move = board.parse_san(move_str)
        if move in board.legal_moves:
            board.push(move)
            result["is_valid"] = True
            result["applied"] = True
            result["new_fen"] = board.fen()
            result["format_detected"] = "SAN"
            return result
    except (ValueError, chess.IllegalMoveError, chess.AmbiguousMoveError) as e:
        result["error"] = f"Invalid move: {str(e)}"

    return result


def get_all_legal_moves(board: chess.Board) -> list:
    """
    Return all legal moves in both UCI and SAN format.

    Args:
        board: Current chess board state

    Returns:
        List of dicts with UCI and SAN representations
    """
    moves = []
    for move in board.legal_moves:
        moves.append({
            "uci": move.uci(),
            "san": board.san(move)
        })
    return moves


# ============================================
# Test: Validate legal and illegal moves
# ============================================
board = chess.Board()  # Starting position

print("Current position (starting):")
print(board)
print(f"FEN: {board.fen()}\n")

# --- Test a valid move ---
print("--- Testing valid move: e2e4 ---")
result = validate_and_apply_move(board, "e2e4")
print(f"  Result: {result}\n")

# --- Test a valid SAN move ---
print("--- Testing valid SAN move: Nf6 ---")
result = validate_and_apply_move(board, "Nf6")
print(f"  Result: {result}\n")

# --- Test an ILLEGAL move (hallucination example) ---
board_test = chess.Board()
print("--- Testing illegal move: e2e5 (skipping a square) ---")
result = validate_and_apply_move(board_test, "e2e5")
print(f"  Result: {result}\n")

# --- Show all legal moves ---
board_fresh = chess.Board()
legal = get_all_legal_moves(board_fresh)
print(f"--- All {len(legal)} legal moves from starting position ---")
for m in legal[:5]:
    print(f"  UCI: {m['uci']}  |  SAN: {m['san']}")
print(f"  ... and {len(legal) - 5} more")
```

```python
#@title 5.0 - Setup & Load Dependencies
# ============================================
# Phase 5: Analysis & Final Report
# Fully standalone - no dependency on other phases
# ============================================

!pip install python-chess cairosvg matplotlib langchain-google-genai langgraph -q

import chess
import chess.svg
import json
import time
import re
import random
from typing import Dict, List, Optional, Any
from collections import Counter
from datetime import datetime
from IPython.display import display, SVG, HTML, Markdown
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.size'] = 12

from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage
from langgraph.graph import StateGraph, END

# --- Config ---
API_KEY = userdata.get('project')
MODEL_NAME = "gemma-3-27b-it"
RATE_LIMIT_DELAY = 2.0

print("âœ… Phase 5 dependencies ready!")
print(f"   Model: {MODEL_NAME}")
```

```python
#@title 5.1 - Rebuild Minimal Agent (Fixed LangGraph State)
# ============================================
# Agent with proper state management
# Compatible with Gemma (no SystemMessage)
# ============================================

SYSTEM_INSTRUCTION = """You are a chess engine. Suggest the best legal move in UCI format.
Only pick from the legal moves list."""

PROMPT_NONE = """You are a chess engine. Suggest the best legal move in UCI format.

FEN: {fen}
Legal moves: {moves}

Respond with ONLY the best move in UCI format (e.g., e2e4), nothing else."""

PROMPT_COT = """You are a chess engine. Analyze this position step by step.

FEN: {fen}
Board:
{board}
Legal moves: {moves}
Turn: {turn}

Analyze:
1. Material balance
2. Threats and tactics
3. Best candidate moves

Then answer on a new line:
BEST_MOVE: <move in UCI format>"""

PROMPT_RETRY = """Move '{move}' is illegal in this position.
Legal moves: {moves}
Pick one legal move. Respond with ONLY the move in UCI format."""


def extract_uci(text: str) -> Optional[str]:
    """Extract UCI move from LLM response."""
    if not text:
        return None
    # Pattern 1: BEST_MOVE: format
    m = re.search(r'BEST_MOVE:\s*([a-h][1-8][a-h][1-8][qrbn]?)', text, re.I)
    if m:
        return m.group(1).lower()
    # Pattern 2: any UCI pattern
    m = re.search(r'\b([a-h][1-8][a-h][1-8][qrbn]?)\b', text)
    if m:
        return m.group(1).lower()
    return None


class Agent:
    """Minimal chess agent with LangGraph - Fixed state management."""

    def __init__(self, mode: str = "none"):
        self.mode = mode
        self.last_call_time = 0

        temp = 0.7 if mode == "sc_cot" else 0.0
        self.llm = ChatGoogleGenerativeAI(
            model=MODEL_NAME,
            google_api_key=API_KEY,
            temperature=temp
        )
        self.graph = self._build()

    def _rate_limit(self):
        """Apply rate limiting."""
        elapsed = time.time() - self.last_call_time
        if elapsed < RATE_LIMIT_DELAY:
            time.sleep(RATE_LIMIT_DELAY - elapsed)
        self.last_call_time = time.time()

    def _ask(self, prompt: str) -> str:
        """Call LLM with rate limiting. No SystemMessage for Gemma."""
        self._rate_limit()
        try:
            resp = self.llm.invoke([HumanMessage(content=prompt)])
            return resp.content
        except Exception as e:
            print(f"\n    âš ï¸ LLM Error: {str(e)[:80]}")
            if "429" in str(e) or "RESOURCE_EXHAUSTED" in str(e):
                print(f"    â³ Rate limited, waiting 60s...")
                time.sleep(60)
                resp = self.llm.invoke([HumanMessage(content=prompt)])
                return resp.content
            raise

    def _build(self):
        """Build LangGraph workflow."""
        g = StateGraph(dict)
        g.add_node("generate", self._generate)
        g.add_node("validate", self._validate)
        g.add_node("retry", self._retry)
        g.add_node("output", self._output)

        g.set_entry_point("generate")
        g.add_edge("generate", "validate")
        g.add_conditional_edges(
            "validate",
            self._should_retry,
            {"retry": "retry", "done": "output"}
        )
        g.add_edge("retry", "validate")
        g.add_edge("output", END)

        return g.compile()

    def _should_retry(self, state: dict) -> str:
        """Decide whether to retry or finish."""
        if state.get("valid", False):
            return "done"
        if state.get("retries", 0) >= 3:
            return "done"
        return "retry"

    def _generate(self, state: dict) -> dict:
        """Generate initial move."""
        fen = state["fen"]
        board = chess.Board(fen)
        legal = [m.uci() for m in board.legal_moves]
        turn = "White" if board.turn else "Black"

        trace = None
        raw_responses = []

        if self.mode == "sc_cot":
            # Self-Consistency: multiple samples
            prompt = PROMPT_COT.format(
                fen=fen, board=str(board),
                moves=", ".join(legal[:20]), turn=turn
            )
            samples = []
            for i in range(5):
                resp = self._ask(prompt)
                raw_responses.append(resp)
                uci = extract_uci(resp)
                if uci:
                    samples.append(uci)

            if samples:
                vote = Counter(samples)
                move = vote.most_common(1)[0][0]
                trace = f"SC-CoT ({len(samples)} valid samples)\nVotes: {dict(vote)}\nSelected: {move}"
            else:
                move = legal[0]
                trace = "SC-CoT: No valid moves extracted, using fallback"

        elif self.mode == "cot":
            prompt = PROMPT_COT.format(
                fen=fen, board=str(board),
                moves=", ".join(legal[:20]), turn=turn
            )
            resp = self._ask(prompt)
            raw_responses.append(resp)
            move = extract_uci(resp)
            if not move:
                move = legal[0]
            trace = resp

        else:  # none
            prompt = PROMPT_NONE.format(
                fen=fen, moves=", ".join(legal[:20])
            )
            resp = self._ask(prompt)
            raw_responses.append(resp)
            move = extract_uci(resp)
            if not move:
                move = legal[0]

        # âœ… Return ALL state keys to prevent KeyError
        return {
            "fen": fen,
            "move": move,
            "legal": legal,
            "trace": trace,
            "raw_responses": raw_responses,
            "retries": 0,
            "valid": False,
            "fallback": False
        }

    def _validate(self, state: dict) -> dict:
        """Validate if move is legal."""
        move = state.get("move", "")
        legal = state.get("legal", [])
        is_valid = move in legal

        # âœ… Return full state with update
        return {
            "fen": state["fen"],
            "move": state["move"],
            "legal": state["legal"],
            "trace": state.get("trace"),
            "raw_responses": state.get("raw_responses", []),
            "retries": state.get("retries", 0),
            "valid": is_valid,
            "fallback": state.get("fallback", False)
        }

    def _retry(self, state: dict) -> dict:
        """Retry with a new prompt."""
        prompt = PROMPT_RETRY.format(
            move=state.get("move", "unknown"),
            moves=", ".join(state.get("legal", [])[:20])
        )
        resp = self._ask(prompt)
        new_move = extract_uci(resp)

        if not new_move:
            new_move = state.get("legal", ["a2a3"])[0]

        raw = state.get("raw_responses", [])
        raw.append(f"RETRY: {resp}")

        # âœ… Return full state with updates
        return {
            "fen": state["fen"],
            "move": new_move,
            "legal": state["legal"],
            "trace": state.get("trace"),
            "raw_responses": raw,
            "retries": state.get("retries", 0) + 1,
            "valid": False,
            "fallback": state.get("fallback", False)
        }

    def _output(self, state: dict) -> dict:
        """Final output - use fallback if needed."""
        if not state.get("valid", False):
            legal = state.get("legal", [])
            fallback_move = random.choice(legal) if legal else "a2a3"
            return {
                "fen": state["fen"],
                "move": fallback_move,
                "legal": state["legal"],
                "trace": state.get("trace"),
                "raw_responses": state.get("raw_responses", []),
                "retries": state.get("retries", 0),
                "valid": True,  # Fallback is always valid
                "fallback": True
            }
        return {
            "fen": state["fen"],
            "move": state["move"],
            "legal": state["legal"],
            "trace": state.get("trace"),
            "raw_responses": state.get("raw_responses", []),
            "retries": state.get("retries", 0),
            "valid": True,
            "fallback": False
        }

    def get_move(self, fen: str) -> dict:
        """Get best move for a position."""
        start = time.time()

        result = self.graph.invoke({"fen": fen})
        elapsed = time.time() - start

        # Apply the move to get new FEN
        board = chess.Board(fen)
        try:
            board.push(chess.Move.from_uci(result["move"]))
            new_fen = board.fen()
        except:
            # Safety fallback
            legal = list(board.legal_moves)
            if legal:
                board.push(legal[0])
                result["move"] = legal[0].uci()
                result["fallback"] = True
            new_fen = board.fen()

        return {
            "move": result["move"],
            "new_fen": new_fen,
            "mode": self.mode,
            "trace": result.get("trace"),
            "raw_responses": result.get("raw_responses", []),
            "fallback": result.get("fallback", False),
            "retries": result.get("retries", 0),
            "time": round(elapsed, 3)
        }


# Quick test
print("Testing agent...")
test_agent = Agent(mode="none")
test_result = test_agent.get_move("rnbqkbnr/pppppppp/8/8/4P3/8/PPPP1PPP/RNBQKBNR b KQkq - 0 1")
print(f"âœ… Agent works! Move: {test_result['move']} | Fallback: {test_result['fallback']}")
```

```python
#@title 5.3 - Statistical Analysis
# ============================================

def compute_stats(results: list) -> dict:
    """Compute comprehensive statistics."""
    stats = {}
    all_phases = ["opening", "tactical", "middlegame", "endgame", "special"]

    for mode in MODES:
        mode_data = [r for r in results if r["mode"] == mode]
        total = len(mode_data)
        if total == 0:
            continue

        valid = sum(1 for r in mode_data if not r.get("fallback", True))
        times = [r["time"] for r in mode_data if r["time"] > 0]
        retries = sum(r.get("retries", 0) for r in mode_data)

        phases = {}
        for phase in all_phases:
            phase_data = [r for r in mode_data if r.get("phase") == phase]
            if phase_data:
                p_valid = sum(1 for r in phase_data if not r.get("fallback", True))
                p_times = [r["time"] for r in phase_data if r["time"] > 0]
                phases[phase] = {
                    "total": len(phase_data),
                    "valid": p_valid,
                    "accuracy": round(p_valid / len(phase_data) * 100, 1),
                    "avg_time": round(sum(p_times) / len(p_times), 3) if p_times else 0
                }

        stats[mode] = {
            "total": total,
            "valid": valid,
            "accuracy": round(valid / total * 100, 1),
            "avg_time": round(sum(times) / len(times), 3) if times else 0,
            "min_time": round(min(times), 3) if times else 0,
            "max_time": round(max(times), 3) if times else 0,
            "total_retries": retries,
            "phases": phases
        }

    return stats


stats = compute_stats(all_results)
all_phases = ["opening", "tactical", "middlegame", "endgame", "special"]

# Print summary
print("="*60)
print("ğŸ“Š STATISTICAL SUMMARY")
print("="*60)
print(f"\n{'Mode':<10} {'Accuracy':<10} {'Avg Time':<10} {'Min':<8} {'Max':<8} {'Retries'}")
print("-"*56)
for mode in MODES:
    s = stats.get(mode, {})
    if s:
        print(f"{mode.upper():<10} {s['accuracy']}%{'':<5} {s['avg_time']}s{'':<4} {s['min_time']}s{'':<2} {s['max_time']}s{'':<2} {s['total_retries']}")

# Phase breakdown
print(f"\n{'Phase':<12} ", end="")
for mode in MODES:
    print(f"{mode.upper():<16}", end="")
print()
print("-"*60)

for phase in all_phases:
    print(f"{phase:<12} ", end="")
    for mode in MODES:
        p = stats.get(mode, {}).get("phases", {}).get(phase, {})
        if p:
            print(f"{p['valid']}/{p['total']} ({p['avg_time']}s){'':<4}", end="")
        else:
            print(f"{'N/A':<16}", end="")
    print()
```

```python
#@title 5.7 - Play a Full Game (Agent vs Agent)
# ============================================

def show_game_board(board, title="", last_move=None):
    """Display board for game."""
    if last_move:
        svg = chess.svg.board(board, lastmove=last_move, size=300)
    else:
        svg = chess.svg.board(board, size=300)
    if title:
        display(HTML(f"<b>{title}</b>"))
    display(SVG(svg))


def play_game(white_mode="cot", black_mode="none", max_moves=20):
    """Play a game between two agent modes."""
    board = chess.Board()
    white = Agent(mode=white_mode)
    black = Agent(mode=black_mode)

    print(f"âš”ï¸ {white_mode.upper()} (White) vs {black_mode.upper()} (Black)")
    print("="*50)

    show_game_board(board, "Starting Position")
    move_log = []

    for i in range(max_moves):
        if board.is_game_over():
            break

        current_agent = white if board.turn else black
        mode_name = white_mode if board.turn else black_mode
        color = "White" if board.turn else "Black"
        move_num = board.fullmove_number

        result = current_agent.get_move(board.fen())
        move_uci = result["move"]

        try:
            move_obj = chess.Move.from_uci(move_uci)
            san = board.san(move_obj)
        except:
            san = move_uci

        prefix = f"{move_num}." if board.turn else f"{move_num}..."
        fb = " (fallback)" if result.get("fallback") else ""
        print(f"{prefix} {san} [{mode_name}]{fb}")

        board.push(chess.Move.from_uci(move_uci))
        move_log.append({
            "num": move_num, "color": color,
            "move": san, "uci": move_uci, "mode": mode_name
        })

        # Show board every 4 half-moves
        if len(move_log) % 4 == 0 or board.is_game_over():
            show_game_board(board, f"After move {len(move_log)}",
                           chess.Move.from_uci(move_uci))

    # Final
    print(f"\n{'='*50}")
    if board.is_checkmate():
        winner = "Black" if board.turn else "White"
        print(f"ğŸ† Checkmate! {winner} wins!")
    elif board.is_stalemate():
        print("ğŸ¤ Stalemate")
    else:
        print(f"ğŸ• Stopped after {len(move_log)} half-moves")

    show_game_board(board, "Final Position")
    return move_log


game_log = play_game("cot", "none", max_moves=20)
```
