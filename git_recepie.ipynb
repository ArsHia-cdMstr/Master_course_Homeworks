### ۱. ساخت توکن دسترسی (GitHub Token)
چون گیت‌هاب دیگر رمز عبور معمولی را برای ترمینال قبول نمی‌کند، باید یک Token بسازید (اگر قبلاً نساخته‌اید):
1. در گیت‌هاب بروید به: **Settings** -> **Developer settings** -> **Personal access tokens** -> **Tokens (classic)**.
2. دکمه **Generate new token** را بزنید.
3. تیک گزینه‌ی **repo** (برای دسترسی کامل به ریپازیتوری‌ها) را بزنید.
4. توکن را کپی کنید و در یک جای امن نگه دارید (چون دیگر آن را نشان نمی‌دهد).
